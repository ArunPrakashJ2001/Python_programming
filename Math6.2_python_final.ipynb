```python
def plotarm(T1,T2,L1,L2):                   # Define function to plot the arm given the lengths and angles as inputs 
    xe=L1*np.cos((np.pi*T1)/180);
    ye=L1*np.sin((np.pi*T1)/180);
    xh=L1*np.cos((np.pi*T1)/180)+L2*np.cos((np.pi*(T1+T2))/180);
    yh=L1*np.sin((np.pi*T1)/180)+L2*np.sin((np.pi*(T1+T2))/180);

    plt.plot([0,xe,xh],[0,ye,yh],"-k")
    plt.plot([0,xe,xh],[0,ye,yh],'o')
    
```

```python
plt.axis([-50,110,-50,110])
plotarm(50,90,T1,T2)
plt.show()
```

$\begin{array}{l}
Code{\ \rm{ and\ Formula\ Correspondence}}\\
{\rm{Th}} \to {\theta _h}\\
Lh \to {L_h}\\
A1 \to {\alpha _1}\\
T10 \to {\theta _1}\\
T20 \to {\theta _2}
\end{array}$

```python
def getT(xh,yh,L1,L2):                           # Define function to find the angles given the current co-ordinates and lengths 
    Th=atan(yh/xh)*(180/np.pi);                  # of the arm 
    Lh=sqrt(xh*xh+yh*yh);
    A1=acos((L1*L1+Lh*Lh-L2*L2)/(2*L1*Lh))*(180/np.pi);
    T10=Th-A1;
    xe=L1*cos(np.pi*T10/180);
    ye=L1*sin(np.pi*T10/180);
    T20=acos((xh-xe)/L2)*(180/np.pi)-T10;
    return T10,T20
```
